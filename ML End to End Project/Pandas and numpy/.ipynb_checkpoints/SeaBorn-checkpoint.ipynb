```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

# Seaborn Plots

```python
tips.describe()
```

```python
tips.info()
```

### Setting the style

```python
plt.style.use('seaborn-v0_8-bright')
sns.set_style('whitegrid')
# Set colorful style and palette globally
sns.set_style('whitegrid')                      # Clean background with grid
sns.set_palette('Set2')                         # Colorful palette (you can try Set1, Pastel1, etc.)
plt.style.use('seaborn-v0_8-colorblind') 
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='day',style='time',size='size')
plt.show()
```

## Distribution Plot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

### Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x="day",y="tip",data=tips)
plt.show()
```

### Line Plot

```python
sns.lineplot(x='size',y='total_bill',data=tips)
```

### Bar Plot

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

### Regression Plot

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

### Violin plot

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

### Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

### Pair Plot

```python
sns.pairplot(data=tips)
```

### Heat Map

```python
numeric_tips = tips.select_dtypes(include='number')
```

```python
sns.heatmap(numeric_tips.corr(), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```

### Relational Plot

```python
sns.relplot(x='total_bill',y='tip',data=tips,col='day',row='time')
```

### Count Plot

```python
tips.sex.value_counts()
```

```python
sns.countplot(x='sex',data=tips,legend='auto')
```

## Subplots

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1) #rows,colomns,plot order
sns.countplot(x='day',data=tips)

plt.subplot(1,2,2)
sns.heatmap(numeric_tips.corr(), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```
