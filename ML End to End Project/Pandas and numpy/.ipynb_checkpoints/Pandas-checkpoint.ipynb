## JOINS

```python
import pandas as pd

left = pd.DataFrame({'CustomerID': [1, 2, 3, 4],'Name': ['Alice', 'Bob', 'Charlie', 'David'],
                             'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
})
left
```

```python
right = pd.DataFrame({
    'OrderID': [101, 102, 103, 104, 105],
    'CustomerID': [3, 1, 1, 5, 2],
    'Amount': [250, 300, 150, 200, 400]
})
right
```

```python
left.merge(right,on='CustomerID',how='inner')
```

```python
T1 = pd.DataFrame({'CustomerID': [1, 2, 3, 4],'Name': ['Alice', 'Bob', 'Charlie', 'David'],
                             'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
})
T1
```

```python
T2 = pd.DataFrame({
    'OrderID': [101, 102, 103, 104, 105],
    'CustomerID': [3, 1, 1, 5, 2],
    'Amount': [250, 300, 150, 200, 400]
})
T2
```

```python
T1.merge(T2,on='CustomerID',how='inner')
```

```python
T1.merge(T2,on='CustomerID',how='outer')
```

```python
pd.concat([T1, T2], axis=0)
```

```python
pd.concat([T1, T2], axis=1)
```

## Text Handling

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Feedback': [
        'Great service!',
        '    Will not @#$ recommend.',
        'Average experience 34.',
        'Loved the 23 ambience.    ',
        'Very quick response.'
    ]
}
```

```python
df =pd.DataFrame(data)
```

```python
df
```

```python
df['lower']=df.Feedback.str.lower()
```

```python
df
```

```python
df['Upper']=df.Feedback.str.upper()
```

```python
df
```

```python
# removing numbers
df['Feedback']=df['Feedback'].str.replace('\d+','',regex=True)
```

```python
df
```

```python
df['Feedback']=df['Feedback'].str.strip()
```

```python
df
```

```python
df['Feedback']=df['Feedback'].str.replace('[@#$]','',regex=True)
```

```python
df
```

```python
df['Feedback']=df['Feedback'].str.title()
```

```python
df
```
