```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Line Plot
- This is mainly used for understanding the changes in a variable across time
- Any data of sequential form can be represented using a line plot

```python
np.random.seed(0)
temperature = np.random.uniform(18,42,size=30)
```

```python
temperature
```

```python
plt.plot(temperature)
```

### Modifying the size of the chart

```python
plt.figure(figsize=(10,5)) #first arg 10 is dimension for X axis and second is Y axis
plt.plot(temperature)
```

```python
plt.figure(figsize=(5,10)) #first arg 10 is dimension for X axis and second is Y axis
plt.plot(temperature)
```

### Multiple line plots

```python
np.random.seed(0)
plt.figure(figsize=(10,5))
summer = np.random.uniform(25,50,size=30)
plt.plot(summer)
winter = np.random.uniform(10,15,size=30)
plt.plot(winter,color='blue')
rainy = np.random.uniform(17,25,size=30)
plt.plot(rainy)

plt.show()
```

### Adding more properties

```python
np.random.seed(0)
plt.figure(figsize=(10,5))
summer = np.random.uniform(25,50,size=30)
plt.plot(summer,color='red',linestyle='dashed')
winter = np.random.uniform(10,15,size=30)
plt.plot(winter,color='blue',lw=5)
rainy = np.random.uniform(17,25,size=30)
plt.plot(rainy,linestyle='dotted')


plt.title('Temperatures aross seasons')
plt.xlabel('30 days period')
plt.ylabel('Temperature')
plt.legend(['summer','winter','rainy'],loc=5) 
plt.show()
```

### Pie Chart

```python
plt.figure(figsize=(8,8))
contries=['india','iraq','srilanka','china']
population=['146','136','20','135']
plt.pie(population,labels=contries,autopct='%1.2f%%')
plt.legend(contries)
plt.show()
```

### Scatter Plot
- To understand the relationship between two variables
- we can see if two varibles are correlated and if there are any clusters

```python
tips= sns.load_dataset('tips')
```

```python
tips
```

```python
plt.scatter(tips.total_bill,tips.tip,color='Red',marker='*')
plt.xlabel('Total_Bill')
plt.ylabel('Tip')
```

```python
tips.corr(numeric_only=True)
```

### Bar Chart

```python
plt.figure(figsize=(7,6))
contries=['india','iraq','srilanka','china']
population=[120,100,25,34]
plt.bar(contries,population)
plt.show()
```

```python
plt.figure(figsize=(7,6))
contries=['india','iraq','srilanka','china']
population=[120,100,25,34]
plt.barh(contries,population,color='red')
plt.show()
```

### Box Plot (Box and Whisker Plot)
- To check outliers in the dataset
- Helps us to understand the data


```python
np.random.seed(1)
junior=np.random.normal(loc=10,scale=2,size=200)
senior=np.random.normal(loc=14,scale=2,size=100)
lead=np.random.uniform(18,25,size=50)
manager=np.random.uniform(24,36,size=20)
vp=np.random.uniform(35,45,size=10)
```

```python
plt.figure(figsize=(10,5))
plt.boxplot([junior,senior,lead,lead,manager,vp],vert=0)
plt.show()
```

### Histogram
- Histogram is used to understand the distribution of a numerical variable

```python
np.random.seed(0)
runs_scored_india = np.random.normal(loc=250, scale= 40, size = 1000)
plt.hist(runs_scored_india,edgecolor='black',bins=20)
```
